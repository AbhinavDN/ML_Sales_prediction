```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("/content/Advertising Budget and Sales.csv")
df.head()
```

```python
df.drop(columns = ["Unnamed: 0"],inplace= True)
```

```python
df
```

```python
X = df.drop(columns = ["Sales"])
X.head()
```

```python
Y = df["Sales"]
Y.head()
```

```python
from  sklearn.model_selection import train_test_split
x_train,x_test,y_train,ytest = train_test_split(X,Y,test_size = 0.2,random_state = 0)
```

```python
x_train.shape
```

```python
from sklearn.linear_model import LinearRegression
linear_reg = LinearRegression()
```

```python
linear_reg.fit(x_train,y_train)
```

```python
linear_reg.intercept_
```

```python
linear_reg.coef_
```

```python
#m1 = 0.04458402, m2 = 0.19649703, m3 = -0.00278146
# Sales  =  (m1*tv) + (m2*radio) + (m3*Newspaper) + c
```

```python
y_pred=linear_reg.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error
mse=mean_squared_error(ytest,y_pred)
mse
```

```python
np.sqrt(mse)
```
